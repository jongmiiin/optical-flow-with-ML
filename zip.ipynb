```python
%%bash

# 2) 변수 설정
DRIVE_ZIP="/content/drive/MyDrive/Capstone/01.원천데이터/TS.zip"
DEST_DIR="/content/drive/MyDrive/Capstone/videos"

# 3) 출력 폴더 생성
mkdir -p "$DEST_DIR"

# 4) ZIP 안의 mp4 목록을 뽑아서 파일에 저장
7z l "$DRIVE_ZIP" | awk '/\.mp4$/ {print $NF}' > mp4_list.txt

# 5) 하나씩 꺼내서 DEST_DIR에 저장
while IFS= read -r mp4path; do
  echo "▶ Extracting: $mp4path"
  7z x "$DRIVE_ZIP" -so "$mp4path" > "$DEST_DIR/$(basename "$mp4path")"
done < mp4_list.txt

echo "✅ All MP4s extracted to $DEST_DIR"
```

```python
%%bash
# 1) 변수 설정
ZIP_PATH="/content/drive/MyDrive/Capstone/01.원천데이터/TS.zip"
DEST_DIR="/content/drive/MyDrive/Capstone/videos"

# 2) 결과 저장 디렉토리 생성
mkdir -p "$DEST_DIR"

# 3) 센서 JSON 전체를 순회하며 mp4 없는 것만 추출
find "/content/TL_unzip/센서" -type f -name "*.json" | while read -r json_file; do
  base="$(basename "$json_file" .json)"
  target="$DEST_DIR/${base}.mp4"

  # mp4 가 이미 있으면 건너뛰기
  if [ -f "$target" ]; then
    continue
  fi

  # ZIP 안의 내부 경로:
  # "/content/TL_unzip/센서" 이후 경로를 뽑아서 "영상/…/*.mp4" 로 바꿔줌
  rel_path="${json_file#/content/TL_unzip/센서}"     # e.g. "/N/N/00781_E_N_C2/00781_E_N_C2.json"
  zip_mp4_path="영상${rel_path%.json}.mp4"             # e.g. "영상/N/N/00781_E_N_C2/00781_E_N_C2.mp4"

  echo "▶ Extracting missing: $zip_mp4_path"
  # 4) 스트림 추출 → flat 파일로 저장
  7z x "$ZIP_PATH" -so "$zip_mp4_path" > "$target"
done

echo "✅ Done. Missing MP4s are now in $DEST_DIR"
```
